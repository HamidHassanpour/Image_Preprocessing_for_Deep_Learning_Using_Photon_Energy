```python
import tensorflow as tf
tf.__version__
from platform import python_version
print(python_version())
(x_train, y_train), (x_test, y_test)=tf.keras.datasets.cifar10.load_data()
print((x_train.shape))
print((y_train.shape))
print((x_test.shape))
print((y_test.shape))
```

```python
x_train_modified=np.zeros((x_train.shape[0],x_train.shape[1],x_train.shape[2],3))
for i in range(x_train.shape[0]):
    x_train_modified[i,:,:,0]=x_train[i,:,:,0]*1.7686833
    x_train_modified[i,:,:,1]=x_train[i,:,:,1]*2.319639
    x_train_modified[i,:,:,2]=x_train[i,:,:,2]*2.618473
    x_train_modified[i,:,:,0]=x_train_modified[i,:,:,0]/np.amax(x_train_modified[i,:,:,0])
    x_train_modified[i,:,:,1]=x_train_modified[i,:,:,1]/np.amax(x_train_modified[i,:,:,1])
    x_train_modified[i,:,:,2]=x_train_modified[i,:,:,2]/np.amax(x_train_modified[i,:,:,2])

x_test_modified=np.zeros((x_test.shape[0],x_test.shape[1],x_test.shape[2],3))
for i in range(x_test.shape[0]):
    x_test_modified[i,:,:,0]=x_test[i,:,:,0]*1.7686833
    x_test_modified[i,:,:,1]=x_test[i,:,:,1]*2.319639
    x_test_modified[i,:,:,2]=x_test[i,:,:,2]*2.618473
    x_test_modified[i,:,:,0]=x_test_modified[i,:,:,0]/np.amax(x_test_modified[i,:,:,0])
    x_test_modified[i,:,:,1]=x_test_modified[i,:,:,1]/np.amax(x_test_modified[i,:,:,1])
    x_test_modified[i,:,:,2]=x_test_modified[i,:,:,2]/np.amax(x_test_modified[i,:,:,2])

print(x_train_modified.shape)
print(x_test_modified.shape)
```

```python
import tensorflow
from tensorflow.keras import layers
from tensorflow.keras import Model
```

```python
from tensorflow.keras.optimizers import RMSprop 

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(32, 32, 3),padding="same"),
    tf.keras.layers.SpatialDropout2D(0.05),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(16, (3,3), activation='relu',padding="same"),
    tf.keras.layers.SpatialDropout2D(0.05),
    tf.keras.layers.MaxPooling2D(2,2), 
    tf.keras.layers.Conv2D(32, (3,3), activation='relu',padding="same"),
    tf.keras.layers.SpatialDropout2D(0.05),    
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu',padding="same"),
    tf.keras.layers.SpatialDropout2D(0.05),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu',padding="same"),
    tf.keras.layers.SpatialDropout2D(0.05),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(), 
    tf.keras.layers.Dense(512, activation='relu'), 
    tf.keras.layers.Dropout(0.05),
    tf.keras.layers.Dense(10, activation='softmax')  
])          

model.compile(optimizer = RMSprop(lr=0.0001),loss = 'categorical_crossentropy',metrics = ['CategoricalAccuracy'])
```

```python
history = model.fit(x=x_train, y=tf.one_hot(y_train[:,0],10),validation_data=(x_test,tf.one_hot(y_test[:,0],10))\
                    ,batch_size=32, epochs=300)
```

```python
history2 = model.fit(x=x_train_modified, y=tf.one_hot(y_train[:,0],10),validation_data=(x_test_modified,tf.one_hot(y_test[:,0],10))\
                    ,batch_size=32, epochs=300)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,10))

categorical_accuracy=(history.history['categorical_accuracy'])
categorical_accuracy2=(history2.history['categorical_accuracy'])
epochs=range(len(categorical_accuracy2))
plt.plot(epochs, categorical_accuracy,'r',label="No Preprocessing")
plt.plot(epochs, categorical_accuracy2, 'g', label="Photon Energy Law Preprocessing" )
plt.legend(loc='upper left', frameon=False)

plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
```

```python
fig=plt.figure(figsize=(10,10))

categorical_accuracy=(history.history['val_categorical_accuracy'])
categorical_accuracy2=(history2.history['val_categorical_accuracy'])
epochs=range(len(categorical_accuracy2))
plt.plot(epochs, categorical_accuracy, 'r',label="No Preprocessing")
plt.plot(epochs, categorical_accuracy2, 'g', label="Photon Energy Law Preprocessing")
plt.legend(loc='upper left', frameon=False)

plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
```

```python
fig=plt.figure(figsize=(10,10))

categorical_accuracy=(history.history['loss'])
categorical_accuracy2=(history2.history['loss'])
epochs=range(len(categorical_accuracy2))
plt.plot(epochs, categorical_accuracy, 'r', label="No Preprocessing")
plt.plot(epochs, categorical_accuracy2, 'g', label="Photon Energy Law Preprocessing")
plt.legend(loc='upper left', frameon=False)

plt.xlabel("Epochs")
plt.ylabel("Training Loss")
```

```python
fig=plt.figure(figsize=(10,10))

categorical_accuracy=(history.history['val_loss'])
categorical_accuracy2=(history2.history['val_loss'])
epochs=range(len(categorical_accuracy2))
plt.plot(epochs, categorical_accuracy, 'r',label="No Preprocessing")
plt.plot(epochs, categorical_accuracy2, 'g', label="Photon Energy Law Preprocessing")
plt.legend(loc='upper left', frameon=False)

plt.xlabel("Epochs")
plt.ylabel("Validation Loss")
```
